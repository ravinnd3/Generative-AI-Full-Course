```python
import numpy as np
import pandas as pd
import gensim
import os
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
import nltk
nltk.download('punkt_tab')
```

```python
import kagglehub

dataset_path = kagglehub.dataset_download("muhammedfathi/game-of-thrones-book-files")

destination_path = "/content/"

# Create the destination directory if it doesn't exist
os.makedirs(destination_path, exist_ok=True)

# Copy the contents of the downloaded dataset to the destination path
# We use shell command for simplicity, you could also use shutil.copytree
!cp -r "{dataset_path}/." "{destination_path}/"

print(f"Dataset copied to: {destination_path}")
```

```python
with open("/content/got1.txt", "r") as file:
    text_content = file.read()

# Display the first 500 characters to confirm content is loaded
print(text_content[:500])
```

```python
story = []
raw_sent = sent_tokenize(text_content)
for sent in raw_sent:
    story.append(simple_preprocess(sent))
```

```python
story
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2
)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model.wv.most_similar('daenerys')
```

```python
model.wv.similarity('arya','sansa')
```
