# Bag of words

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'text':['people watch youtube','youtube watch youtube','people write comment','youtube write comment'],'output':[1,1,0,0]})
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
bow = cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

```python
bow.toarray()
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
print(bow[2].toarray())
print(bow[3].toarray())
```

```python
# to pass into model
x = bow.toarray()
y = df['output']
```

# N-grams

```python
# Bi-grams
cv = CountVectorizer(ngram_range=(2,2))

df = pd.DataFrame({'text':['people watch youtube','youtube watch youtube','people write comment','youtube write comment'],'output':[1,1,0,0]})
df
```

```python
bow = cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
print(bow[2].toarray())
```

```python
#Tri gram
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(3,3))
```

```python
bow = cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
print(bow[2].toarray())
```

# TF-IDF (Term frequency- Inverse document frequency)

```python
df = pd.DataFrame({'text':['people watch youtube','youtube watch youtube','people write comment','youtube write comment'],'output':[1,1,0,0]})
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer()
```

```python
arr = tf.fit_transform(df['text']).toarray()
```

```python
arr
```

```python
print(tf.idf_)
```
